# Digital Coin Flip and 1D Random Walk
This notebook demonstrates two simple simulations:
- A digital coin flip
- A 1D random walk

## Section 1: Digital Coin Flip
Simulate flipping a coin and display the result.

```python
import random

# Digital coin flip function
def coin_flip():
    return "Heads" if random.choice([True, False]) else "Tails"

# Flip the coin and display the result
result = coin_flip()
print(f"Coin flip result: {result}")
```

## Section 2: 1D Random Walk
Simulate a 1D random walk for a given number of steps and plot the position over time.

```python
import matplotlib.pyplot as plt

# 1D random walk function
def random_walk(steps=100):
    position = 0
    positions = [position]
    for _ in range(steps):
        step = random.choice([-1, 1])
        position += step
        positions.append(position)
    return positions

# Simulate and plot the random walk
steps = 100
walk = random_walk(steps)
plt.plot(range(steps + 1), walk)
plt.xlabel('Step')
plt.ylabel('Position')
plt.title('1D Random Walk')
plt.show()
```
